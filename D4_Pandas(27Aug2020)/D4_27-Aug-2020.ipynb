## Pandas Series

```python
# list into pandas series
a = pd.Series([1,2,3,4,5],index = range(10,15))
type(a)
a
```

```python
s=pd.Series(range(8))
s
```

```python
# dict in series
import numpy as np
dic1 = {'K1' : 'V1','K2' : 'V2', 'K3' : 'V3',"k4":np.nan}
s2 = pd.Series(dic1)
s2
```

```python
dic1 = {'K1' : 'V1','K2' : 'V2', 'K3' : 'V3',"k4":np.nan}
s2 = pd.Series(dic1,index = ["K1","K3","P"])
s2
```

```python
# explicitly indexing
s1=pd.Series((1,2,3,4,5,6))
s1.index=['a','b','c','d','e','f']
s1
```

###  Note : make sure the length of list elements always equal to the data elements

```python
# create series with power of index 0 - 10  
se = pd.Series(np.arange(10)**2,index = range(1,10))
se
```

```python
se = pd.Series(np.arange(1,11)**2,index = range(1,11))
se
```

```python
s=pd.Series(np.arange(0,11)**2) # default index starts from 0
s
```

## Pandas Dataframe
**2-D data**
* Dataframe is the most widely used data-structure in data analysis. It is a table with rows and columns, with rows having an index and columns having meaningful names.

* Creating dataframes from dictionaries

* There are various ways of creating dataframes, such as creating them from dictionaries, JSON objects, reading from txt, CSV     files, etc.

* *One of the main purposes of pandas is to deal with tabular data, i.e. data that comes from tables or spreadsheets.*

```python
di1 = {
    "employee" : ["a","b","c","d"],
    "group" : ["ec","cs","mech","civil"]
}

di2 = {
    "employee" : ["a","c","d","b"],
    "age" : [23,24,17,25]
    
}
df1 = pd.DataFrame(di1)
df2 = pd.DataFrame(di2)
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
di1 = {
    "employee" : ["a","b","c","d","l"],
    "group" : ["ec","cs","mech","civil","eee"]
}
di2 = {
    "employee" : ["a","c","d","b"],
    "age" : [23,24,17,25]    
}
df1 = pd.DataFrame(di1)
df2 = pd.DataFrame(di2)
df1
```

```python
df2
```

```python
pd.merge(df1,df2)  # it returns only common rows in both df's
```

```python
df_con = pd.merge(df1,df2,how="left")
df_con.columns = ["emp","grp","age"]
df_con
```

```python
pd.merge(df1,df2,how="right")
```
