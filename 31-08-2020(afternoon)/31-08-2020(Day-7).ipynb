### Line plot:

* Joining all the points using a line.

```python
x = [1,2,3,4]
y = [1,2,3,4]

plt.plot(x,y) #### plotting graph for x and y values.
plt.show()
```

```python
plt.plot(x,y,'m')
plt.show()
```

```python
plt.plot(x,y,'g')
plt.xlabel("x-axis")  ## giving name for x-axis
plt.ylabel("y-axis")## giving name for y-axis
plt.title("sample line plot") ## title of the plot
plt.show()
```

```python
import numpy as np
x = np.linspace(0,5,50)
x
```

```python
plt.plot(x,x,color = "g",label = "liner graph")
plt.plot(x,x**2,color = 'r',label = "x**2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("line plot with multiple plotting")
plt.legend()
plt.show()
```

### Subplots:

* subplot is a function used to plot mutliple plots at a time.
* Subplots(number_of rows,number_of columns)


```python
### creating subplot:

plt.subplots(2,2)
plt.show()
```

```python
fig,a = plt.subplots(2,2)
x = np.arange(1,5) 

a[0][0].plot(x,x)
a[0][0].set_title("linear")
a[0][1].plot(x,x**2)
a[0][1].set_title("Square")
a[1][0].plot(x,np.exp(x))
a[1][0].set_title("Exponential")
a[1][1].plot(x,np.log(x))
a[1][1].set_title("Logarithm")
plt.tight_layout()
plt.show()
```

```python
a1 = plt.subplot2grid((3,3),(0,0),colspan = 2)
a2 = plt.subplot2grid((3,3),(0,2),rowspan = 3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan = 2,colspan = 2)
x = np.arange(1,10)
a2.plot(x,x**2)
a2.set_title("square")
a1.plot(x,np.log(x))
a1.set_title("log")
a3.plot(x,np.sqrt(x))
a3.set_title("square root")
plt.tight_layout()
plt.show()
```
