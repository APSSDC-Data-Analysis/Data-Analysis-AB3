# Data Visualization using Seaborn

    * Seaborn is a Python data visualization library based on matplotlib
    * It provides a high-level interface for drawing attractive and informative statistical graphics
    * Applications - used in visualising data in Machine learning, data Science 
[seaborn official_docs](https://seaborn.pydata.org/)

    * This function provides an interface to many (though not all) of the possible ways you can generate colors in seaborn

2. Plotting with categorical data

    **Categorical scatterplots:**

        * stripplot() (with kind="strip"; the default)

        * swarmplot() (with kind="swarm")

    **Categorical distribution plots:**

        * boxplot() (with kind="box")

        * violinplot() (with kind="violin")

# Boxplot
* distrubution along with extreme values

```python
sns.catplot(y = "species", x = "petal_length",data = iris,kind= "box")
plt.show()
```

```python
sns.catplot(data = iris,kind= "box",orient = "h") # ver/hor
plt.show()
```

## Violinplots
* Distrubutions

```python
sns.catplot(y = "species", x = "petal_length",data = iris,kind= "violin")
plt.show()
```

```python
sns.violinplot(x = iris.species , y = iris.sepal_width) 
```

```python
help(sns.violinplot)
```

## Joint Plot
* Combination of two plots


```python
sns.jointplot(x = "petal_length", y = "sepal_length",data = iris)
plt.show()
```

```python
sns.jointplot(x = "petal_length", y = "sepal_length",data = iris,kind = "reg")
plt.show()
```

```python
help(sns.jointplot)
```

```python
sns.jointplot(x = "petal_length", y = "sepal_length",data = iris,kind = "kde")
plt.show()
# represents density of data 
```

```python
sns.jointplot(x = "petal_length", y = "sepal_length",data = iris,kind = "hex")
plt.show()
```

## Pairplot
* It's used to plot multiple plots at a time for entire data

```python
sns.pairplot(iris, hue = "species",diag_kind= "kde",kind = "scatter")
plt.show()
```

```python
sns.pairplot(iris, hue = "species",diag_kind= "kde",kind = "scatter",markers = ["d","^","*"])
plt.show()
```

## Heatmap
* Heatmap is a way show some sort od matrix plot
* correlation between each n every labels
* numerical data

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips_cor = tips.corr()
tips_cor
```

```python
plt.figure(figsize  = (10,8))
sns.heatmap(tips_cor)
```

```python
sns.heatmap(tips_cor,annot = True)
```

```python
tips.shape
```

```python
sns.heatmap(tips_cor,annot = True,cmap = "plasma") # color map
```

```python
help(sns.heatmap)
```
